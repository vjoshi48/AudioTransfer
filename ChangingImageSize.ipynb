```python
import glob
import torch
import numpy
from pdb import set_trace
import cv2

image_paths_list = []

image_paths_list.append(glob.glob('Images/train/trap/*'))
image_paths_list.append(glob.glob('Images/train/rock/*'))
image_paths_list.append(glob.glob('Images/val/rock/*'))
image_paths_list.append(glob.glob('Images/val/trap/*'))

for i in range(len(image_paths_list)):
  for j in range(len(image_paths_list[i])):
    image = cv2.imread(image_paths_list[i][j], cv2.IMREAD_GRAYSCALE).astype('float32')
    image = image[:, :1300]
    for k in range(2):
      if k == 1:
        cv2.imwrite(image_paths_list[i][j], image[:, :650])
      else:
        image_name = image_paths_list[i][j][:-4] + str(k) + '.png'
        num1 = (650*(k+1))
        num2 = 650 + (650*(k+1))
        try:
          cv2.imwrite(image_name, image[:, num1:num2])
        except:
          set_trace()
```
